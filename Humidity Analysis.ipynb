## according to the information provided in the website from where the data is downloaded-->the figures showing -200 are missing values


```python
#counting the number of times "-200" has appeared in the data

d.isin([-200]).sum(axis=0)
```

## Convert all "-200" values to NaN

```python
d=d.replace(to_replace=-200,value=np.NaN)
```

```python
d
```

```python
d.isnull().sum()     #gives the number of null values in the columns
```

### All the (-200) values have been converted to null values

```python
d.describe()
```

```python
# fill the Null values with the mean of the rows     
d=d.fillna(d.mean())
```
